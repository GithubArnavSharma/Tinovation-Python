```python
#Import neccessary modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

print("Imported successfully")
```

```python
cancer = load_breast_cancer()
print(cancer)
```

```python
#Get the X and y arrays from the breast cancer dataset
X = cancer['data']
y = cancer['target']
print(X.shape)
print(y.shape)
```

```python
#Visualize all 30 of the normalized attributes 
cancer_atr = np.array(X)
rows = 5
columns = 6
atr = 0
fig, ax = plt.subplots(rows, columns, figsize=(10,10))
for i in range(rows):
    for j in range(columns):
        sns.histplot(ax=ax[i][j], x=cancer_atr[:,atr], kde=True)
        ax[i][j].set_xlabel(cancer.feature_names[atr])
        atr += 1
plt.tight_layout()
plt.show()
```

```python
#Visualize the target
sns.countplot(x=y, palette='rocket')
plt.xlabel('Has Breast Cancer')
plt.show()
```

```python
#Get the training and testing arrays for the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
#Train the support vector machine model on the training data
model = SVC()
model.fit(X_train, y_train)
print("SVM trained")
```

```python
#Test the SVM on the testing data
print(model.predict(X_test))
print(y_test)
print(model.score(X_test, y_test))
```
