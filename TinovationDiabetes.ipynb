```python
#Import neccessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
print("Imported Successfully")
```

```python
#Load and print dataset information
diabetes = load_diabetes()
print(diabetes)
```

```python
#Visualize all 10 of the normalized attributes 
diabetes_atr = np.array(diabetes.data)
rows = 5
columns = 2
atr = 0
fig, ax = plt.subplots(rows, columns)
for i in range(rows):
    for j in range(columns):
        sns.histplot(ax=ax[i][j], x=diabetes_atr[:,atr], kde=True)
        ax[i][j].set_xlabel(diabetes.feature_names[atr])
        atr += 1
fig.tight_layout()
plt.show()
```

```python
#Retrieve and visualize the frequency of the target variable for the diabetes dataset
diabetes_target = pd.DataFrame(diabetes.target, columns=['Target'])
sns.histplot(diabetes_target, kde=True)
plt.xlabel('Disease Progression')
plt.show()
```

```python
#Convert the dataset into x and y arrays 
X = np.array(diabetes['data'])
y = np.array(diabetes['target'])
print(X.shape)
print(y.shape)
```

```python
#Split the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#Create the Linear Regression model and fit it onto the training data
model = LinearRegression()
model.fit(X_train, y_train)
print("Model trained")
```

```python
#Get the predictions for the test variable and print out a quantitative measure of the error
y_pred = model.predict(X_test)
print(y_pred)
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```
